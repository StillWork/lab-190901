# 랜덤신호 곱과 자신의 곱

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.random.uniform(-1, 1, 300)
plt.plot(range(300), x)
```

```python
sum_xy = []
sum_xx = []
for i in range(60):
    x = np.random.uniform(-1, 1, 100)
    y = np.random.uniform(-1, 1, 100)
    sum_xy.append((x*y).sum())
    sum_xx.append((x*x).sum())
plt.plot(range(60), sum_xy)
plt.plot(range(60), sum_xx, c ='r')
```
